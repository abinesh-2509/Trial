```python
# prompt: forecast the price of bitcoin using statistical methods like moving averages for the next 6 months and plot it

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, timedelta

# Download historical Bitcoin price data
bitcoin = yf.Ticker("BTC-USD")
end_date = date.today()
start_date = end_date - timedelta(days=365 * 2)  # Get data for the last 2 years
data = bitcoin.history(start=start_date, end=end_date)

# Calculate moving averages (e.g., 50-day and 200-day)
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()

# Forecast the next 6 months using the last available moving averages
forecast_days = 6 * 30  # Approximation of 6 months in days
forecast_dates = pd.date_range(start=end_date, periods=forecast_days + 1, freq='D')[1:]
forecast_data = pd.DataFrame({'Date': forecast_dates, 'Forecast': [data['MA50'][-1]] * forecast_days})
forecast_data = forecast_data.set_index('Date')


# Combine historical and forecast data
combined_data = pd.concat([data, forecast_data])

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Actual Price')
plt.plot(combined_data['MA50'], label='50-Day MA')
plt.plot(combined_data['MA200'], label='200-Day MA')
plt.plot(forecast_data['Forecast'], label='6-Month Forecast', linestyle='--') # Highlight the forecast
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Price Forecast')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: Draw pacman

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataframe with 2 columns and 10 rows
data = {'col1': np.random.rand(10), 'col2': np.random.rand(10)}
df = pd.DataFrame(data)

# Drawing Pacman
# Create a figure and an axes object
fig, ax = plt.subplots()

# Define the center and radius of the circle
center_x = 0
center_y = 0
radius = 1

# Create the circle
circle = plt.Circle((center_x, center_y), radius, color='yellow')

# Add the circle to the axes
ax.add_patch(circle)

# Create the mouth (a wedge)
mouth_start_angle = 30
mouth_end_angle = 150

# Create the wedge
wedge = plt.Wedge((center_x, center_y), radius, mouth_start_angle, mouth_end_angle, color='white')
ax.add_patch(wedge)


# Set the aspect ratio of the plot to 'equal'
ax.set_aspect('equal')

# Set the limits of the plot
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])

# Add the title
ax.set_title("Pacman")

# Show the plot
plt.show()
```
